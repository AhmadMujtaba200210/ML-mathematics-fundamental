###**Solving Linear Systems**


Equations:
                  
                  y=3x

                -5x+2y=2
Rearranged 2nd equation to Isolate y:

                y= (2+5x)/2

                2y=2+5x

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import tensorflow as tf
```

```python
x= np.linspace(-10,10,1000) # start finish endpoints
```

```python
y1=3*x
```

```python
y2=1+(5*x)/2
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0,3])
ax.set_ylim([0,8])
ax.plot(x,y1,c="green")
ax.plot(x,y2,c="brown")
plt.axvline(x=2,color="purple",linestyle="--")
plt.axhline(y=6,color="purple",linestyle="--")
```

##Elimination Example

In the Elimination example, the two equations in the system are:

                        2x - 3y = 15
                        4x + 10y = 14
Both equations can be rearranged to isolate y. Starting with first equation:

                        -3y = 15 -2x
                        y = (15-2x)/-3 = -5 + 2x/3
Then for the second equation:

                        4x + 10y = 14
                        2x + 5y = 7
                        5y = 7 - 2x
                        y = (7-2x)/5

##Eigen Vectors

Affine Transformation via Matrix Multiplication

```python
import matplotlib.pyplot as plt
```

```python
v = np.array([3,1])
v
```

```python
def plot_vectors(vectors,colors):
  plt.figure()
  plt.axvline(x=0,color='lightgray')
  plt.axhline(y=0,color='lightgray')

  for i in range(len(vectors)):
    x = np.concatenate([[0,0],vectors[i]])
    plt.quiver([x[0]],[x[1]],[x[2]],[x[3]],
               angles='xy',scale_units='xy',scale=1,color=colors[i])
```

we will plot a graph in which a vector is stretches from 0 to 3 on x axis and 0 to 1 on y axis i.e. array = [3,1]

```python
plot_vectors([v],['lightblue'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

Applying matrix ( multiplying matrix )

```python
I = np.array([[1,0],[0,1]])
I
```

```python
product = np.dot(I,v)
product
```

```python
plot_vectors([product],['darkblue'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

Lets multiply a matrix with such matrix that will flip the matrix on x axis

```python
E = np.array([[1,0],[0,-1]])
E
```

```python
Ev = np.dot(E,v)
Ev
```

```python
plot_vectors([v,Ev],['darkblue','darkred'])
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.title('Flipping matrix along x axis')
```

```python
F = np.array([[-1,0],[0,1]])
F
```

```python
Fv = np.dot(F,v)
Fv
```

```python
plot_vectors([Fv,v],['blue','red'])
plt.xlim(-4,4)
plt.ylim(-1,5)
plt.title("Flippling Matrix along y axis")
```

So, this is affine transformation where we change in geometry such as direction and angles, where flipping a matrix is also a part of geometry.

A single matrix can apply multiple affine transformation such as flipping a matrix on x or y axis and then rotate to 45 degree

```python
A = np.array([[-1,4],[2,-2]])
A
```

```python
Av = np.dot(A,v)
Av
```

```python
plot_vectors([Av,v],['blue','red'])
plt.xlim(-4,4)
plt.ylim(-1,5)
```
