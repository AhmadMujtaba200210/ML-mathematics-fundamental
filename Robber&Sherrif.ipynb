```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Plotting Time

```python
t=np.linspace(0,40) #start, stop
```

2. Distance of robber and sherrif.

```python
d_r = 2.5*t 
d_s = 3*(t-5)
```

3. Creating the plot

```python
fig,ax=plt.subplots()
plt.title('Robber vs Sherrif at Bank Robbery')
plt.xlabel('Time --->')
plt.ylabel('Distance --->')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,d_r,c='green')
ax.plot(t,d_s,c='blue')
plt.axvline(x=30,color='purple',linestyle='--')
plt.axhline(y=75,color='purple',linestyle='--')
```
